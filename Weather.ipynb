```python
import pandas as pd
weather = pd.read_csv("weatherAUS.csv")
weather
```

```python
weather.drop(labels = ["Date"], axis = 1, inplace = True)
```

## Preproccessing

### Dealing with null values

```python
#number of null values in each column
weather.isnull().sum()
```

```python
#if a column has more than 4000 null value we drop it
labels=["Evaporation", "Sunshine", "WindGustDir","WindGustSpeed","WindDir9am","WindDir3pm","Humidity3pm","Pressure9am","Pressure3pm","Cloud9am","Cloud3pm"]
weather.drop(labels = labels, axis = 1, inplace = True)
weather
```

```python
#fill null values with median
weather["MinTemp"] = weather["MinTemp"].fillna(weather["MinTemp"].median())
weather["MaxTemp"] = weather["MaxTemp"].fillna(weather["MaxTemp"].median())
weather["Rainfall"] = weather["Rainfall"].fillna(weather["Rainfall"].median())
weather["WindSpeed9am"] = weather["WindSpeed9am"].fillna(weather["WindSpeed9am"].median())
weather["WindSpeed3pm"] = weather["WindSpeed3pm"].fillna(weather["WindSpeed3pm"].median())
weather["Humidity9am"] = weather["Humidity9am"].fillna(weather["Humidity9am"].median())
weather["Temp9am"] = weather["Temp9am"].fillna(weather["Temp9am"].median())
weather["Temp3pm"] = weather["Temp3pm"].fillna(weather["Temp3pm"].median())
weather
```

```python
#fill null values in "RainToday" with mode (most frequent value)
weather["RainToday"] = weather["RainToday"].fillna(value = weather["RainToday"].mode()[0])
weather
```

```python
#delete rows in "RainTomorrow" that has null value
weather.drop(weather.index[(weather["RainTomorrow"].isnull())] , axis = 0, inplace = True)
weather
```

### Dealing with categorical values

```python
import sklearn.preprocessing as skp
LE = skp.LabelEncoder()
weather["Location"] = LE.fit_transform(weather["Location"])
weather["RainToday"].replace({"Yes": 1,"No": 0}, inplace = True)
weather["RainTomorrow"].replace({"Yes": 1,"No": 0}, inplace = True)
weather
```

### normalization

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit_transform(weather)
```

### Split data into train, validation and test

```python
from sklearn.model_selection import train_test_split
train,validation = train_test_split(weather,train_size=0.8, test_size=0.2, random_state=0)
train,test=train_test_split(train ,train_size=0.8, test_size=0.2, random_state=0)
```

```python
#split X,y
import warnings
warnings.filterwarnings('ignore')
y_train=train["RainTomorrow"]
train.drop(labels =["RainTomorrow"], axis = 1, inplace = True)
X_train=train
y_validation=validation["RainTomorrow"]
validation.drop(labels =["RainTomorrow"], axis = 1, inplace = True)
X_validation=validation
y_test=test["RainTomorrow"]
test.drop(labels =["RainTomorrow"], axis = 1, inplace = True)
X_test=test
```

## Create multilayer perceptron 

```python
from sklearn.neural_network import MLPClassifier
wModel = MLPClassifier(hidden_layer_sizes=(7, 4), activation='relu',solver='adam', alpha=0.001,batch_size=200, random_state=1)
```

## Training

```python
wModel.fit(X_train, y_train)
```

## Evaluate model with accuracy

```python
from sklearn.metrics import accuracy_score
ytrain_pred=wModel.predict(X_train)
ytest_pred=wModel.predict(X_test)
train_acc=accuracy_score(y_train,ytrain_pred)
test_acc=accuracy_score(y_test,ytest_pred)
print("accuracy of train: ",train_acc,"accuracy of test: ",test_acc)
```

```python
import pandas as pd
te = pd.read_excel("temprature.xlsx")
te
```

```python
te.drop(labels =["Date"], axis = 1, inplace = True)
te
```

## Split data into train, validation and test

```python
from sklearn.model_selection import train_test_split
train,validation = train_test_split(te,train_size=0.8, test_size=0.2, random_state=0)
train,test=train_test_split(train ,train_size=0.8, test_size=0.2, random_state=0)
```

## Creating Data (last 5 days)

```python
from keras.preprocessing.sequence import TimeseriesGenerator
train_ge=TimeseriesGenerator(train.values,train.values,length=5, batch_size=20)
validation_ge=TimeseriesGenerator(validation.values,validation.values,length=5, batch_size=20)
test_ge=TimeseriesGenerator(test.values,test.values,length=5, batch_size=20)
```

## Creating LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras import backend as K

def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 

model = Sequential()
model.add(LSTM(units=50,return_sequences=True))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss = root_mean_squared_error)
```

## Training

```python
model.fit(train_ge, epochs=20, verbose=1)
```

## Computing loss for train and test

```python
train_loss=model.evaluate(train_ge)
test_loss=model.evaluate(test_ge)
print("rmse loss of train: ",train_loss,"and rmse loss of test: ",test_loss)
```

## predict 3 first samples of test
